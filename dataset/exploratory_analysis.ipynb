```python
import os
import glob
import numpy as np
import pandas as pd
import pathlib
```

```python
dataset_path =  os.path.abspath(".")
dataset_path
```

```python
# looking for the images (they all are .jpg)
globpath = f"{dataset_path}/**/*.jpg"

# generation of an array with all the pictures
pictures = np.hstack([np.array(glob.glob(globpath, recursive = True))])
pictures[:10]
```

```python
def get_annotation_file(input_path):
    '''
    takes image filename and returns annotation filename
    e.g. data/Images/IMG_XXX.jpg
         data/Annotations/IMG_XXX.txt
    INPUTS:
        input_path: path to image file
    OUTPUTS:
        output_path: path to annotation file
    '''
    if type(input_path) == str:
        input_path = pathlib.Path(input_path)
    output_name = input_path.stem + '.txt'
    output_path = input_path.parent.parent / 'Annotations' / output_name
    if output_path.exists():
        return output_path
    else:
        raise FileNotFoundError(f'{output_path} does not exist')

def get_annotation_list(image_list):
    '''
    For input list of images, return a dataframe with label information.
    dictionary keys: image file, class (string), x, y, h, w 
    '''
    full_dataframe = None
    
    for jpg_path in image_list:
        annotation_path = get_annotation_file(jpg_path)
        data = pd.read_csv(
            annotation_path, 
            delimiter=' ', 
            names=['class','x','y','w','h']
        )
        data['filename'] = jpg_path
        
        if full_dataframe is None:
            full_dataframe = data.copy()
        else:
            full_dataframe = full_dataframe.append(data, ignore_index=True)   
    return full_dataframe
```

```python
full_image_list = glob.glob(globpath, recursive = True)
df = get_annotation_list(full_image_list)
```

```python
# all images have been resized to 640x640 pixels
df['width'] = 640.
df['height'] = 640.

# calculate xmin, xmax, multiply by box width (640)
df['xmin'] = df['width']*(df['x'] - (df['w']/2.0))
df['xmax'] = df['width']*(df['x'] + (df['w']/2.0))
# calculate ymin, ymax, multiply by box height (640)
df['ymin'] = df['height']*(df['y'] - (df['w']/2.0))
df['ymax'] = df['height']*(df['y'] + (df['w']/2.0))

# If you want integers, simply use df['ymin'] = df['ymin'].map(int)
df['ymin'] = df['ymin'].map(lambda x: '%3.2f' % x)
df['ymax'] = df['ymax'].map(lambda x: '%3.2f' % x)
df['xmin'] = df['xmin'].map(lambda x: '%3.2f' % x)
df['xmax'] = df['xmax'].map(lambda x: '%3.2f' % x)
```
